```python
# Normal Distribution (Bell-Curve Plot)
# Continuous variables (- infinity to infinity)--- wide range of values(fractional, decimal vlaues included)
#--- Defines a sample size
np.random.seed(0) 
random_data = np.random.normal(loc = 50000, scale = 10000, size = 1000)
random_data
```

```python
sns.histplot(random_data, kde=True)
```

```python
random_data.mean()
```

```python
mu, std = stats.norm.fit(random_data)
```

```python
mu
```

```python
std
```

```python
stats.norm.cdf(60000, mu, std) - stats.norm.cdf(40000, mu, std)
```

```python
# Gamma Distribution (Mostly used for Time functions as a distribution)
# More convienient for Time based values
# It is a positive distribution and doesn't except negative values

time_to_purchase = np.random.gamma(shape=2, scale=1, size=1000)
time_to_purchase
```

```python
#Statistical testing
# Parametric tests only use Normal Distribution, Non-Parametric test (Exponential tests do not use parametric tests)

from scipy.stats import ttest_ind, mannwhitneyu

group_a = np.random.normal(30000, 5000, 1000)
group_b = np.random.normal(35000, 5000, 1000)
```

```python
t_stat, p_value = ttest_ind(group_a, group_b)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/delinai/schulich_ds1_2024/main/Datasets/class_5_data.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.head(10)
```

```python
df['Customer_ID'].nunique()
```

```python
plt.hist(df.groupby('Customer_ID')['Purchase_Amount'].count())
```

```python
# user level data
users = df.groupby('Customer_ID')[['Campaign_Exposure', 'Purchase_Amount', 'Premium_Product']].agg(
    {'Campaign_Exposure': 'max', 'Purchase_Amount': 'mean', 'Premium_Product':['mean', 'count']}
)
```

```python
users
```

```python
exp = users[users['Campaign_Exposure']['max'] == 1]
not_exp = users[users['Campaign_Exposure']['max'] == 0]
```

```python
# alternative compares the values between the two variables (+ indicates supporting data based on the argument)
#t_stat, p_value = ttest_ind(not_exp['Purchase_Amount'], exp['Purchase_Amount'], alternative='less')

t_stat, p_value = ttest_ind(not_exp['Purchase_Amount'], exp['Purchase_Amount'])
```

```python
p_value
```

```python
(exp['Purchase_Amount'].mean() / not_exp['Purchase_Amount'].mean()) - 1
```

```python
# Non-parametric test

u_stat, p_value = mannwhitneyu(not_exp['Premium_Product']['count'], exp['Premium_Product']['count'], alternative = 'less')
print(p_value)
```

```python
u_stat, p_value = mannwhitneyu(not_exp['Premium_Product']['mean'], exp['Premium_Product']['mean'], alternative = 'less')
print(p_value)
```
