```python
# Unsupervised Learning
# We do not have an output variable
# No final prediction but is used to find combinations and patterns within the data
```

```python
# Examples of Unsupervised Learning:
# 1. Customer Segementation
# 2. Gene and DNA Matching/Verification
# 3. Vector Simiarity
# 4. Anamoly  Detection (Fraud, Machine Overdrive, Computing)
```

```python
# Clustering Types:

# 1. Partitioning (K-means): Data is divided into K-groups
# Overlap between the clusters, might be difficult to separate out the clusters
# The algorithm will help identify the number of clusters
# A classification model can be used post to identify which cluster is the best for future data
# Categorical data cannot be used in k-means as the data is not continuous data
# K-means and DBSCAN need to have continuous data
# Eg: Customer segementation (you have to personally identify the different features to group the individuals to a particular cluster)
# Average and complete linkage is best for cleanly separated data
# Ward Linkage is best for noisy data

# 2. Density-based clusters: 
# How dense is a particular section of the dataset
# Data points could be heavily clustered or data could be spread out unevely 
# Good to detect outliers and deal with outliers
# DBSCAN Cluster is more efficient while handling outliers
# Mean-shift: Various the centroid to find the optimal point to divide the data into appropriate clusters

# 3. Distribution-model clusters
# Based on the data points belonging to a certain distribution

# 4. Hierarchical clusters:
# Eg: Social media websites: Facebook, instagram (Relation between different people)
# Eg: Gene Mapping

# 5. Fuzzy Mapping:
# Based on a belonging to a certain groups, eg: slangs and abreviations being categorized to the main word

# 6. Vectorization
# Eg: Words, Sentences can be converted to vectors and compare to different segements
# Eg: Combining various apps on a mobile to a particular embedded vector and then use it for comparison with other vectors
```

```python
# K-Means Evaluation:
# Only K-Means requires a evaluation metric or method to verify if the clusters are optimal or not
# Others methods do not have evaluation metrics

# 1. The Elbow Method:
# Two metrics are used: Distortion and Inertia and plot them on a graph
# This help to decide the optimal level of clusters and find out the best concentration to determine optimal clusters
# This can be determined if there is a very small drop between the values of k(number of clusters)
# Test upto 10 clusters to verify the optimal number and choose the best cluster based on score of cluster that hasn't reduced significantly

# 2.Silhouette Analysis 
# Useful to analyze the distance between the different clusters (how the data is spread)
# Can check if there is no overlap
# Scores range betweeen -1 and 1; -1: means overlapped clusters and 1: far apart clusters
# Scores closer to 1 are great meaning the clusters are separated optimally

 
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=300, centers = 4, cluster_std = 0.60, random_state=0)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.title('Synthetic Data')
plt.show()
```

```python
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(range(1,11), inertia)
plt.title('Elbow Method')
plt.xlabel('Num Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], c = 'red')
plt.title('4 clusters')
plt.show()
```

```python
labels
```

```python
kmeans = KMeans(n_clusters=6)
kmeans.fit(X)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
plt.scatter(X[:,0], X[:,1], c=labels)
plt.scatter(centers[:,0], centers[:,1], c = 'red')
plt.title('6 clusters')
plt.show()
```

```python
import pandas as pd
df = pd.DataFrame(X)
df['label'] = labels
df
```

```python
sil_score = silhouette_score(X, labels)
print(sil_score)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

Z = linkage(X, method='ward')
plt.figure(figsize=(10,5))
dendrogram(Z)
plt.title('hierarchical clustering')
plt.show()
```

```python
# t = 5 means we are creating 5 clusters
clusters = fcluster(Z, t=5, criterion = 'maxclust')
```

```python
clusters
```

```python
plt.scatter(X[:,0], X[:,1], c=clusters)
plt.title('5 clusters')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

```python
if len(set(dbscan_labels)) > 1:
    dbscan_sil_score = silhouette_score(X, dbscan_labels)
    print(dbscan_sil_score)
```

```python
plt.scatter(X[:,0], X[:,1], c = dbscan_labels)
plt.title('DBSCAN Clustering')
plt.show()
```

```python
df['hour'] = df.index.map(lambda x: x[11:])
df['hour']
```

```python
datagroup = df.groupby('hour')
avg_curves = datagroup.agg('mean')
avg_curves.shape
```

```python
avg_curves_norm = avg_curves / (avg_curves.mean())
avg_curves_norm
```

```python
avg_curves_norm[['MT_001','MT_002','MT_369','MT_370']].plot()
```

```python
import numpy as np
X = avg_curves_norm.copy()
X = np.array(X.T)
X
```

```python
def getInertia(X,kmeans):
    inertia = 0
    for J in range(len(X)):
        inertia = inertia + np.linalg.norm(X[J] - kmeans.cluster_centers_[kmeans.labels_[J]])**2
    return inertia
```

```python
inertia_1 = []
silhouette_avgs = []
max_K = 15
for k in np.arange(1,max_K+1):
    kmeans=KMeans(n_clusters=k, random_state=0).fit(X)
    inertia_1 = inertia_1 + [kmeans.inertia_]
    if k>1:
        silhouette_avgs = silhouette_avgs + [silhouette_score(X,kmeans.labels_)]
```

```python
plt.figure(figsize = (20,5))
plt.subplot(1,2,1)
plt.title('Inertia: sum square distances to closest cluster')
plt.plot(np.arange(1,max_K+1), inertia_1)
plt.xticks(np.arange(1,max_K+1))
plt.xlabel('num_clusters')
plt.grid()
```

```python
K = 6
kmeans = KMeans(n_clusters=K, random_state=0).fit(X)
for c in range(K):
    avg_curves_norm.T[kmeans.labels_ == c].T.plot(legend=False)
    plt.plot(kmeans.cluster_centers_[c], linewidth=10, color='orange')
    plt.title('Cluster '+str(c)+": centroid (orange)")
    plt.show()
```
