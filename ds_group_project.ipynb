```python
# Define products, categories, prices, and weights
PRODUCTS = [
    (random.randint(10000, 99999), "Eggs(0.5dozen)", "Dairy", 0.30, 4.5, random.randint(1, 4)),  # Weight in kg (1 egg ~ 50g), Price per egg
    (random.randint(10000, 99999), "Eggs(1dozen)", "Dairy", 0.60, 7.5, random.randint(1, 3)),
    (random.randint(10000, 99999), "Eggs(1.5dozen)", "Dairy", 0.90, 11.25, random.randint(1, 3)),
    (random.randint(10000, 99999), "Brown Bread", "Bakery", 0.5, 5, random.randint(1, 4)),
    (random.randint(10000, 99999), "White Bread", "Bakery", 0.5, 4, random.randint(1, 4)),
    (random.randint(10000, 99999), "Biscuits", "Snacks", 0.35, 4.5, random.randint(1, 5)),  # 1 pack ~ 350g
    (random.randint(10000, 99999), "Strawberry Jam", "Snacks", 0.35, 4.75, random.randint(1, 2)),  # 1 jar ~ 350g
    (random.randint(10000, 99999), "Peanut Butter", "Dairy", 0.5, 4.5, random.randint(1, 2)),  # 1 jar ~ 500g
    (random.randint(10000, 99999), "Tortilla(6pck)", "Bakery", 0.30, 5, random.randint(1, 4)),  # 1 pack ~ 300g(50g/tortilla)
    (random.randint(10000, 99999), "Tortilla(10pck)", "Bakery", 0.50, 7.5, random.randint(1, 2)),
    (random.randint(10000, 99999), "Rice(2kg)", "Grains", 2.0, 6.0, random.randint(1, 3)),  # 1kg ~ $3
    (random.randint(10000, 99999), "Rice(4kg)", "Grains", 4.0, 11.0, random.randint(1, 3)),
    (random.randint(10000, 99999), "Butter(350g)", "Dairy", 0.35, 4.75, random.randint(1, 3)),  # 1 block ~ 350g
    (random.randint(10000, 99999), "Butter(550g)", "Dairy", 0.55, 6, random.randint(1, 4)),
    (random.randint(10000, 99999), "Cheese(14pckts)", "Dairy", 0.35, 6.5, random.randint(1, 4)),
    (random.randint(10000, 99999), "Shredded Cheese(350g)", "Dairy", 0.35, 7, random.randint(1, 4)),
    (random.randint(10000, 99999), "Shredded Cheese(620g)", "Dairy", 0.62, 12.5, random.randint(1, 3)),
    (random.randint(10000, 99999), "Pasta", "Grains", 0.5, 4.5, random.randint(1, 4)),  # 1 pack ~ 500g
    (random.randint(10000, 99999), "Chocolates", "Snacks", 0.25, 5, random.randint(1, 5)),  # 1 bar ~ 250g
    (random.randint(10000, 99999), "Apples(5pck)", "Fruits", 0.8, 5, random.randint(1, 4)),
    (random.randint(10000, 99999), "Bananas(6pck)", "Fruits", 0.95, 4.5, random.randint(1, 4)), 
    (random.randint(10000, 99999), "Onions(5pck)", "Vegetables", 0.85, 4, random.randint(1, 4)),
    (random.randint(10000, 99999), "Mushrooms(350g)", "Vegetables", 0.35, 3.75, random.randint(1, 4)),
    (random.randint(10000, 99999), "Mushrooms(500g)", "Vegetables", 0.50, 5.25, random.randint(1, 3)),
    (random.randint(10000, 99999), "Chicken", "Meat", 1.0, 16.0, random.randint(1, 4)),  # 1kg
    (random.randint(10000, 99999), "Beef", "Meat", 1.0, 18.0, random.randint(1, 4)),  # 1kg
    (random.randint(10000, 99999), "Pork", "Meat", 1.0, 17.0, random.randint(1, 4)),  # 1kg
    (random.randint(10000, 99999), "Cookies", "Snacks", 0.35, 4.25, random.randint(1, 5)),  # 1 pack ~ 350g
    (random.randint(10000, 99999), "Cupcakes(6pck)", "Snacks", 0.60, 5.5, random.randint(1, 4)),  # 1 cupcake ~ 100g
    (random.randint(10000, 99999), "Cereal", "Snacks", 0.5, 4.0, random.randint(1, 4)),  # 1 box ~ 500g
    (random.randint(10000, 99999), "Tomato Ketchup", "Sauce", 0.75, 5.5, random.randint(1, 3)),  # 1 bottle ~ 750g
    (random.randint(10000, 99999), "Mustard", "Sauce", 0.35, 4.5, random.randint(1, 3)),  # 1 bottle ~ 350g
    (random.randint(10000, 99999), "Mayonnaise", "Sauce", 0.5, 6, random.randint(1, 2)),  # 1 bottle ~ 500g
    (random.randint(10000, 99999), "Hot Sauce", "Sauce", 0.35, 4.0, random.randint(1, 3))  # 1 bottle ~ 350g
]

CUSTOMERS = [
(random.randint(100000, 999999),"Leslie Pitts", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Jeffrey Jordan", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Cory Simpson", fake.date_between(start_date='-5y', end_date='-1y')),
(random.randint(100000, 999999),"Emily Bryant", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Kim Scott", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Michael Romero", fake.date_between(start_date='-5y', end_date='-1y')),
(random.randint(100000, 999999),"Kylie Davis", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Eric Patterson", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Alison Williams", fake.date_between(start_date='-5y', end_date='-1y')),
(random.randint(100000, 999999),"Elizabeth Brown", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Richard Barnes", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Micheal Case", fake.date_between(start_date='-5y', end_date='-1y')),
(random.randint(100000, 999999),"James Saunders", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"John Castillo", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Sandra Gibson", fake.date_between(start_date='-5y', end_date='-1y')),
(random.randint(100000, 999999),"Beth King", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Michael Morales", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Matthew Gordon", fake.date_between(start_date='-5y', end_date='-1y')),
(random.randint(100000, 999999),"Muskan Sekhri", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Neeraj Nagula", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Abhinav Vig", fake.date_between(start_date='-5y', end_date='-1y')),
(random.randint(100000, 999999),"Selvin Furtado", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Xinyue Wang", fake.date_between(start_date='-5y', end_date='-1y')),(random.randint(100000, 999999),"Delina Ivanova", fake.date_between(start_date='-5y', end_date='-1y')),
]

# Generate synthetic data
def generate_synthetic_data(n=1000):
    data = []
    last_transaction_dates = {}
    historical_amounts = {} 
    for _ in range(n):
        products = random.choice(PRODUCTS)
        customers = random.choice(CUSTOMERS)
        product_id, product_name, product_category, avg_weight, product_price, product_count = products
        customer_id, customer_name, customer_start_date = customers
        total_amount = round(product_price * product_count, 2)

        # Update historical amount
        if customer_name not in historical_amounts:
            # For the first transaction, historical_amount is equal to total_amount
            historical_amount = total_amount
        else:
            # For subsequent transactions, set historical_amount equal to the previous transaction's historical_amount
            historical_amount = historical_amounts[customer_name]
        
        # Store historical amount for the next transaction
        historical_amounts[customer_name] = total_amount

        transaction_id = fake.uuid4()
        transaction_amount = total_amount
        transaction_time = fake.date_time_this_year()

        # Initialize Last transaction date
        if customer_id not in last_transaction_dates:
            last_transaction_dates[customer_id] = transaction_time
        else:
            # Update the last transaction date 
            last_transaction_dates[customer_id] = max(last_transaction_dates[customer_id], transaction_time)       

        data.append({
            "product_id": product_id,
            "product_name": product_name,
            "product_price": product_price,
            "average_weight_kg": avg_weight,
            "product_category": product_category,
            "customer_id": customer_id,
            "customer_name": customer_name,
            "product_count": product_count,
            "total_amount": total_amount,
            "historical_amount": historical_amount,
            "customer_start_date": customer_start_date,
            "last_transaction_date": last_transaction_dates[customer_id].date(),
            "transaction_id": transaction_id,
            "transaction_amount": transaction_amount,
            "transaction_time": transaction_time,
        })
    return pd.DataFrame(data)
```

```python
# Function to generate a barcode image
def generate_barcode(product_id):
    ean = barcode.get_barcode_class('ean13')
    barcode_img = ean(str(product_id).zfill(12), writer=ImageWriter())
    buffer = BytesIO()
    barcode_img.write(buffer)
    return buffer.getvalue()
```
